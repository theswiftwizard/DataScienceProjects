```python
import pandas as pd
import matplotlib.pyplot as plt

netflix_df = pd.read_csv("./netflix_data.csv")
# print(netflix_df.iloc[1, :])

# print the first five rows    
print(netflix_df.iloc[0:5])
```

```python
netflix_df_movies_only = netflix_df[netflix_df['type'] == 'Movie']
# print(netflix_df_movies_only[:4])

# subsetting netflix_df_movies_only so that title country genre release_year and duration are preserved
netflix_movies_col_subset = netflix_df_movies_only[['title', 'country', 'genre', 'release_year', 'duration']]
print(netflix_movies_col_subset[:5])
```

```python
# Creating a scatter plot
# Create a figure and increase the figure size
fig = plt.figure(figsize=(12,8))
x = netflix_movies_col_subset['release_year']
y = netflix_movies_col_subset['duration']
plt.scatter(x, y)
plt.title("Movie Duration by Year of Release")
plt.show()
```

```python
short_movies = netflix_movies_col_subset[netflix_movies_col_subset['duration'] < 60]
# print(type(short_movies))
print(short_movies[:20])
```

```python
colors = []
for g in netflix_movies_col_subset['genre']:
    if g == 'Children':
        colors.append("red")
    elif g == "Documentaries":
        colors.append("blue")
    elif g == "Stand-Up":
        colors.append("green")
    else:
        colors.append("black")
    
        
print(colors[:10])
```

```python
# Set the figure style and initalize a new figure
plt.style.use('fivethirtyeight')

fig = plt.figure(figsize = (12, 8))
plt.scatter(netflix_movies_col_subset['release_year'], netflix_movies_col_subset['duration'], c=colors)
plt.xlabel("Release year")
plt.ylabel("Duration (min)")
plt.title("Movie duration by year of release")
plt.show()
```
